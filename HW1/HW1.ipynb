# Часть I: О теореме Байеса

Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!

1. Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.

2. Затем выходит адвокат, объясняет ошибки прокурора и берёт слово сам: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Какие вероятности он оценил, а какие должен был оценить?


__Постановка задачи:__
    
- Событие $А$ - подсудимый виновен, обозначим $P(A) = p$.

Априорно вероятность $P(A)$ следует понимать с точки зрения классического определения вероятности, как $P(A) = \frac{1}{n}$, где $n$ - число подозреваемых.

- Событие $B$ - у подсудимого редкая группа крови, $P(B) = \frac{1}{100}$.

Для определения степени вины подозреваемого целесообразно рассмотреть вероятность, что он виновен, при условии что у него редкая группа крови, т.е. апостериорную вероятность $P( A | B)$.

__Байесовский вывод:__

По т. Байеса

$
P(A | B) = \frac{P(B | A) P(A)}{P(B)}
$

Из условия, что на месте убийства найдена кровь редкой группы, которая явно, с вероятностью 1, принадлежит настоящему убийце,  
следует, что правдоподобие $P(B | A) = 1$, тогда

$
P(A | B) = \frac{P(A)}{P(B)} = 100p.
$

Заметим, что события $A$ и $B$ не являются независимыми, действительно $0 < P(B) < 1$ и

$
P(A B) = P(A | B) P(B) = P(A) \ne P(A) P(B).
$

__Апостериорная вероятность в 100 раз больше априорной__, именно на это обстоятельство мог обратить внимание прокурор.

Адвокат же обязан был указать, что __совместная вероятность обоих событий зависит только от априорной степени виновности__   
и не зависит от того, что именно у этого подсудимого кровь редкой группы, поэтому с одинаковой апостериорной вероятностью  
в качестве подсудимого может рассматриваться любой подозреваемый (и вообще любой человек) с редкой группой крови.  

Ошибка адвоката в том, что он допустил независимость обоих событий и неправильно оценил вероятность совместного события,  
при этом его оценка оказалась в 100 раз ниже чем есть на самом деле

$
P(A B) \, "\!=\!" \: P(A) P(B) = \frac{p}{100}.
$

В чем ошибка прокурора и какую веростность он рассмотрел, тут порассуждаем более детально

- $P$ {посудимый виновен и у него редкая группа крови}
$P(A B) = P(A)$
- $P$ {посудимый виновен и у него обычная группа крови}
$P(A \overline B) = 0$
- $P$ {посудимый не виновен и у него редкая группа крови}
$P(\overline A B) = P(B) - P(A B) = P(B) - P(A)$
- $P$ {посудимый не виновен и у него обычная группа крови}
$
P(\overline A \: \overline B) = P(\overline B | \overline A) P(\overline A) 
 = (1 - P(B | \overline A)) P(\overline A) \quad(*)
$

Посчитаем условную вероятность того, что подсудимого редкая группа крови, при условии что он не виновен
$
P(B | \overline A) = \frac{P(B \overline A)} {P(\overline A)} = \frac{P(B) - P(A B)}{P(\overline A)} =
\frac{P(B) - P(A)}{1 - P(A)},
$

отметим, что $P(B | \overline A) \ne P(B)$, сами события, как и их отрицания, независимы  
(интуиция такая, что среди всех подозреваемых с редкой группой крови, все же есть 1 - виновный).


Тода в (*) получим  
$
P(\overline A \: \overline B) = \left(1 - \frac{P(B) - P(A)}{1 - P(A)} \right) (1 - P(A))
= 1 - P(B)
$  

Легко проверить, что сумма вероятностей всех совместных событий равна 1.

Предполагаю, что прокурор рассмотрел условную вероятность $P(B | \overline A)$, при этом допустил, что события независимы, т.е.  
$
P(B | \overline A) = \frac{P(B \overline A)} {P(\overline A)} = \frac{P(B) - P(A B)}{P(\overline A)} 
\, "\!=\!" \frac{P(B) - P(A) P(B)}{1 - P(A)} = P(B) = \frac{1}{100}.
$

Далее, прокурор сделал ошибочныю цепочку выводов  
$
P(B | \overline A) = \frac{1}{100} \, "\!\implies\!" P(B | A) = \frac{99}{100} \, "\!\implies\!" P(A) = \frac{99}{100}.
$


### Вернёмся к тесту на страшную болезнь из первой лекции.

Напомню, что по предположениям задачи болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны.  
Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%. 
Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт положительный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).  
Помогите медикам: постройте зависимости апостериорной вероятности болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?


```python
import numpy as np
from matplotlib import pyplot as plt

def posterior_positive(fp, fn, prior):
    return prior * (1 - fn) / (fp * (1 - prior) + (1 - fn) * prior)

def posterior_negative(fp, fn, prior):
        return prior * fn / ((1 - fp) * (1 - prior) + fn * prior)

p = .01
pp = posterior_positive(0.05, 0.05, p)
pn = posterior_negative(0.05, 0.05, p)
display(f'prior: {p}')
display(f'posterior positive: {pp}')
display(f'posterior negative: {pn}')
display(f'positive/negative:  {pp / pn}')
```

```python
fig = plt.figure(figsize=(16, 6))

p = .01
fx = .05
xs = np.linspace(0, .05, 100)

plt.subplot(221)
plt.plot(xs, posterior_positive(xs, fx, p), label='Positive test   | FP error')
plt.plot(xs, posterior_positive(fx, xs, p), label='Positive test   | FN error')
plt.plot(xs, posterior_negative(xs, fx, p), lw=5.0, label='Negative test | FP error')
plt.plot(xs, posterior_negative(fx, xs, p), ls='--',label='Negative test | FN error')
plt.title(f'Disease posterior with prior {p*100}%')
plt.xlabel('error rate (FP or FN)')
plt.ylabel('posterior')
plt.legend()

plt.subplot(222)
plt.plot(xs, posterior_negative(xs, fx, p), lw=5.0, c='green', label='Negative test | FP error')
plt.plot(xs, posterior_negative(fx, xs, p), ls='--',c='red', label='Negative test | FN error')
plt.title(f'Disease posterior with prior {p*100}% for Negative test')
plt.xlabel('error rate (FP or FN)')
plt.ylabel('posterior')
plt.legend()

p = .1
plt.subplot(223)
plt.plot(xs, posterior_positive(xs, fx, p), label='Positive test   | FP error')
plt.plot(xs, posterior_positive(fx, xs, p), label='Positive test   | FN error')
plt.plot(xs, posterior_negative(xs, fx, p), lw=5.0, label='Negative test | FP error')
plt.plot(xs, posterior_negative(fx, xs, p), ls='--',label='Negative test | FN error')
plt.title(f'Disease posterior with prior {p*100}%')
plt.xlabel('error rate (FP or FN)')
plt.ylabel('posterior')
plt.legend()

plt.subplot(224)
plt.plot(xs, posterior_negative(xs, fx, p), lw=5.0, c='green', label='Negative test | FP error')
plt.plot(xs, posterior_negative(fx, xs, p), ls='--',c='red', label='Negative test | FN error')
plt.title(f'Disease posterior with prior {p*100}% for Negative test')
plt.xlabel('error rate (FP or FN)')
plt.ylabel('posterior')
plt.legend()

plt.tight_layout()
plt.show()
```

__Вывод:__

- С математической точки зрения график posterior(FP | FN) лежит выше posterior(FN | FP), надо уменьшать FP.

- С человеческой точки зрения, лучше триста раз FN, и дообследование, чем один раз FP, надо уменьшать FN.
 
Врачам следует принимать во внимание совокупность признаков, таких как смертность, заразность,  
длительность повторной диагностики и лечения, также не стоит забывать стоймость всех необходимых процедур,  
и в каждом конкретном случае принимать обоснованное решение.

```python
ACTUAL_COLUMNS = ['date','total_cases','new_cases']
COUNTRY_NAME = 'Russia'
DATE_MIN = '2020-03-03'
TRAIN_DAYS = 50
TEST_DAYS = 10

def get_country_df(country, date_min=DATE_MIN, columns=ACTUAL_COLUMNS):
    mask = (df_full.location == country) & (df_full.date >= DATE_MIN) 
    df = df_full[mask][columns].copy()
    
    get_days = lambda x: (pd.to_datetime(x) - pd.to_datetime(DATE_MIN)).days
    df['days'] = df['date'].apply(get_days)
    
    df['total_cases_log'] = np.log(df['total_cases'])
    df['new_cases_log'] = np.log(df['new_cases'])

    return df

'get_country_df defined'
```

```python
def get_train_test(df, train_days=TRAIN_DAYS, test_days=TEST_DAYS):
    mask = (df.days >= 0) & (df.days < TRAIN_DAYS)
    train = df.loc[mask,:].copy()
    mask = (df.days >= TRAIN_DAYS) & (df.days < TRAIN_DAYS + TEST_DAYS)
    test = df.loc[mask,:].copy()
    return train, test

'get_train_test defined'
```

### 2.a Модель с экспоненциальным ростом

Используя линейную регрессию, обучите модель с экспоненциальным ростом числа заболевших: 

y ~ exp(линейная функция от x), где x — номер текущего дня.

```python
class ExponentialGrowingModel:
    def __init__(self):
        self.w = np.array([0., 0.])
        
    def coef(self):
        return self.w[::-1]
    
    def fit(self, x, y):
        self.w = np.polyfit(x, np.log(y), len(self.w) - 1)
        
    def predict(self, x):
        return np.exp(x * self.w[0] + self.w[1])

'ExponentialGrowingModel'
```

```python
exp_model_total_cases = ExponentialGrowingModel()
exp_model_total_cases.fit(train['days'], train['total_cases'])
train['total_cases_exp_pred'] = exp_model_total_cases.predict(train['days'])
train['total_cases_exp_log'] = np.log(train['total_cases_exp_pred'])
display(f'total_cases_model: w={exp_model_total_cases.coef()}')

exp_model_new_cases = ExponentialGrowingModel()
exp_model_new_cases.fit(train['days'], train['new_cases'])
train['new_cases_exp_pred'] = exp_model_new_cases.predict(train['days'])
train['new_cases_exp_log'] = np.log(train['new_cases_exp_pred'])
display(f'new_cases_model:   w={exp_model_new_cases.coef()}')

'Fitting done'
```

### 2.b Апостериорное распределение параметров модели

Найдите апостериорное распределение параметров этой модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оцените, исходя из вашей же максимальной апостериорной модели (это фактически первый шаг эмпирического Байеса).

```python
def get_likelihood(px, py, sigma=.5):
    return lambda x : np.exp(-(x[0] + x[1]*px - py) ** 2) / (2 * sigma * np.sqrt(2.*np.pi))

def bayesian_update(mu, sigma, x, y, sigma_noise=.25):
    x_matrix = np.concatenate((np.ones((x.shape[0], 1)), x), axis=1)
    sigma_n = np.linalg.inv(np.linalg.inv(sigma)+ (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), x_matrix) )
    mu_n = np.matmul(sigma_n, np.matmul(np.linalg.inv(sigma), np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), np.array(y) ) )
    return mu_n, sigma_n
```

__Байесовский вывод:__

- noise: $\quad \: \mathcal{N}(0, \sigma)$

- prior: $\quad \: \: p(\theta) = \mathcal{N}(\theta|0, \sigma_0^2 I)$

- posterior: $p(\theta|D) = \mathcal{N}(\theta| \mu_N , \Sigma_N)$

- model: $\quad \mu_N = \Sigma_N\left(\Sigma^{-1}_0\mu_0+\frac{1}{\sigma^2}X^Tln(y)\right)$  
$\quad \quad \quad \: \: \Sigma_N = \left(\Sigma^{-1}_0+\frac{1}{\sigma^2}X^TX\right)^{-1}$




```python
true_mu = exp_model_total_cases.coef()
display(('true_mu', true_mu))

cur_mu, cur_sigma = np.array([0, 0]), 2*np.array([[1, 0], [0, 1]])

fig = plt.figure(figsize=(12, 3))

N = 250
X = np.linspace(-3, 3, N)
Y = np.linspace(-3, 3, N)
X, Y = np.meshgrid(X, Y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
  
def myplot_heatmap(Z, true_mu, title):
    plt.title(title)
    plt.pcolormesh(X, Y, Z, cmap=plt.cm.jet,shading='auto')
    plt.scatter(true_mu[0], true_mu[1], marker='*', s=120)

# Prior
Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
plt.subplot(141)
myplot_heatmap(Z, true_mu, 'Prior')

# Likelihood
i = 1
px = train['days'].to_numpy()[i]
py = train['total_cases_log'].to_numpy()[i]
cur_likelihood = get_likelihood(px, py)
Z = np.array([[ cur_likelihood(pos[i, j]) for j in range(pos.shape[1])] for i in range(pos.shape[0])])
plt.subplot(142)
myplot_heatmap(Z, true_mu, 'Likelihood')

# Empirical Bayes step
px = np.asarray([px]).reshape(-1, 1)
py = np.asarray([py])
cur_mu, cur_sigma = bayesian_update(cur_mu, cur_sigma, px, py)
Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
plt.subplot(143)
myplot_heatmap(Z, true_mu, 'Empirical Bayes')

plt.show()
```

Empirical Bayes estimation

```python
def make_estimation(x, target, pred, mu_prior=None, sigma_prior=None):
    mu = mu_prior
    if mu_prior is None:
        mu = np.array([0., 0.])
        
    sigma = sigma_prior
    if sigma_prior is None:
        sigma = 2*np.array([[1., 0.], [0., 1.]])

    sigma_noise = np.var(target - pred)
    
    mu, sigma = bayesian_update(mu, sigma, x, target, sigma_noise)
    return mu, sigma


def plot_estimation(mu_true, mu_est, sigma_est, label):
    fig = plt.figure(figsize=(6, 6))

    X = np.linspace(mu_true[0]-0.2, mu_true[0]+0.2, N)
    Y = np.linspace(mu_true[1]-0.02, mu_true[1]+0.02, N)
    X, Y = np.meshgrid(X, Y)

    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X
    pos[:, :, 1] = Y

    Z = multivariate_normal.pdf(pos, mean=mu_est, cov=sigma_est)

    plt.pcolormesh(X, Y, Z, cmap=plt.cm.jet,shading='auto')
    plt.scatter(mu_true[0], mu_true[1], marker='*', s=120)
    plt.title(label)
    plt.xlim((mu_true[0]-0.2, mu_true[0]+0.2))
    plt.ylim((mu_true[1]-0.02, mu_true[1]+0.02))
    plt.show()
```

```python
x = train['days'].to_numpy().reshape(-1, 1)
target = train['total_cases_log'].to_numpy()
pred = train['total_cases_exp_log'].to_numpy()

mu_true = exp_model_total_cases.coef()
mu_est, sigma_est = make_estimation(x, target, pred)

plot_estimation(mu_true, mu_est, sigma_est, 'Total cases estimation')
display(('Expected mu: ', mu_true))
display(('Estimated mu:', mu_est))

mu_total_cases = mu_est
sigma_total_cases = sigma_est
```

### 3. Модель с сигмоидальной функцией

Предсказания экспоненциальной модели наверняка получились грустными. Но это, конечно, чересчур пессимистично — экспоненциальный рост в природе никак не может продолжаться вечно. Кривая общего числа заболевших во время эпидемии в реальности имеет сигмоидальный вид: после начальной фазы экспоненциального роста неизбежно происходит насыщение. В качестве конкретной формы такой сигмоиды давайте возьмём форму функции распределения для гауссиана.  
Естественно, в нашем случае сигмоида стремится не к единице, т.е. константа перед интегралом может быть произвольной (и её можно внести в экспоненту), а в экспоненте под интегралом может быть произвольная квадратичная функция от t.

$$ \Phi(x) = \int_{-\infty}^x e^{at^2+bt + c} dt \quad(*)$$

__Предложите способ обучать параметры такой сигмоидальной функции.__

Общее число заболевших на заданную дату - это сумма новых заболевших от начала измерения до заданной даты.  
$$total\_cases(n) = \sum_{i=0}^n new\_cases(i)$$

Значаения new_cases(t) можно интерпретировать как приращения total_cases(t), при переходе от дискретных значений к непрерывным получим  
$$total\_cases'(n) = new\_cases(x)$$

Для сигмоидальной функции модели (*) производная равна подинтегральному выражению.  
Обучим экспоненциальную модель с функцией $e^{at^2+bt + c}$ на значениях new_cases.  
Для предсказания total_cases просуммируем все прогнозы экспоненциальной модели до заданной даты.

```python
class SigmoidGrowingModel:
    def __init__(self):
        self.w = np.array([0., 0., 0.])
        
    def coef(self):
        return self.w[::-1]
    
    def fit(self, x, y):
        self.w = np.polyfit(x, np.log(y), len(self.w) - 1)
        
    def predict(self, x):
        return np.exp(x**2 * self.w[0] + x * self.w[1] + self.w[2])
    
    def predict_cumsum(self, x):
        return np.cumsum(self.predict(x))

'SigmoidGrowingModel'
```

```python
def plot_sigma_new_preds(mu, sigma, date_pred, date_from=DATE_MIN, n_samples=100):
    days = (pd.to_datetime(date_pred) - pd.to_datetime(date_from)).days + 1
    ws = np.random.multivariate_normal(mu, sigma, n_samples)
    preds = np.exp(ws[:, 0] + ws[:, 1] * days + ws[:, 2] * (days**2) )
    display(f'New cases:   {int(preds.mean())}')
    plt.hist(preds, bins=30)
    plt.title(f'New cases in Russia to {date_pred}')
    
def plot_sigma_total_preds(mu, sigma, date_pred, date_from=DATE_MIN, n_samples=100):
    days = (pd.to_datetime(date_pred) - pd.to_datetime(date_from)).days + 1
    ws = np.random.multivariate_normal(mu, sigma, n_samples)
    preds = np.zeros((days, ws.shape[0]))
    for i in range(days):
        ws = np.random.multivariate_normal(mu, sigma, n_samples)
        preds[i,:] = np.exp(ws[:, 0] + ws[:, 1] * i + ws[:, 2] * (i**2) ).mean()
    total_cases = preds.cumsum()
    display(f'Total cases in Russia expected to {date_pred}')
    display(f'Mean:        {int(total_cases.mean())}')
    display(f'Pessimistic: {int(np.percentile(total_cases, 90))}')
    display(f'Optimistic:  {int(np.percentile(total_cases, 10))}')
    plt.hist(total_cases, bins=30)
    plt.title(f'Total cases in Russia to {date_pred}')
```
